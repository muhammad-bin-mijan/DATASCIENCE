````python
#Arithmetic Operators:
val_1,val_2 = 10,20
print("Addition:")
print(val_1+val_2)
print("Subtraction:")
print(val_1-val_2)
print("Multiplication:")
print(val_1*val_2)
print("Division:")
print(val_1/val_2)
print("Floor Division:")
print(val_1//val_2)
print("Modulus:")
print(val_1%val_2)
print("Exponentiation:")
print(2**3)
````

````python
#Assignment operator
x = 5
y = 10
x+=5 # x = x + 5
print(x)
x-=5
print(x)
x*=5
print(x)
x/=5 # It will always return a float value.
print(x)
x//=5
print(x)
x%=5
print(x)
````

````python
#comparison operators
# It will return bool type always
print(5>3)
print(5==6)
````

````python
#Logical Operators
print(6>8 and 6>7)
print(5>1 or 5>6)
print(not 5>1)
````

````python
#identity operators  is , is not 
x=[1,2,3]
y=[1,2,3]
z=x
print(x is z)
print(x is y)
print(x==y)
print(x is not y)
````

````python
# membership operator
# Checks if an element is in a sequence
li = ['banana','cherry','mango']
print('banana' in li)
print('apple'not in li)
````

````python
#Swapping
x = 10
y = 5
x,y = y,x
print(x,y)
````
