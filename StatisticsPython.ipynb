```python
import numpy as np
print(np.mean(ages))
print(np.median(ages))
#mean = 37.36363636363637
```

```python
#If I add a value say 100 in ages then the mean will change
ages = [23, 24, 32, 45, 12, 43, 67, 45, 32, 56, 32,100]
```

```python
print(np.mean(ages))
# mean = 42.583333333333336
# Change in mean implies that mean is affected by the large value added at last (100) 
# Mean changed by 5 
# 100 is an outlier here in age dataset
# Mean is not resistant and not  stable
```

```python
#Median does not change because of outliers because it is the middle of the sorted elements and it will not 
# change so much by outliers 
# So it is resistant and stable
print(np.median(ages))
```

```python
#Another way to find mean ,median 
import statistics
print(statistics.mean(ages))
print(statistics.median(ages))
print(statistics.mode(ages))
```

#**KEY TAKEAWAYS**

#**5 Number Summary**

```python
q1,q3 = np.percentile(ages,[25,75])
```

```python
print(f"First Quartile: {q1} , third Quartile: {q3}")
```

```python
# To check outliers(lower fence ,higher fence)
IQR=q3-q1
lower_fence = q1 - ( 1.5*IQR)
higher_fence = q3 + (1.5*IQR)
print(f"lower fence : {lower_fence} , higher fence :{higher_fence}")
```

#**Any value outside of [lower fence,higher fence] is outlier**
