```python
#List 
li = [1,2,3,4,5,5]
#list can have duplicate values
print(li)
```

```python
#Python add list items
#append()
li.append(6)# appends one item at a time at the end and modifies in place
print(li)
#Even if I append a list it will be added as an item ,nested list
li.append([6,7,8,9])
print(li)
#insert an item  insert(index,item) it also modifies in place.Also if I insert an item at an index it does not replace the item 
#that was in that index previously
li.insert(8,0)
print(li)
#extend() adds each element of another iterable individually
li_1 = [1,2,3,4]
li_1.extend(range(5,10))#in place modifies
print(li_1)
```
