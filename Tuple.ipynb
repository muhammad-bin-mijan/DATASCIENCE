```python
#Tuple
# 📌 Tuple Properties in Python
# 🔑 Core Characteristics
# - Ordered → Elements maintain the order in which they are defined.
# - Immutable → Cannot be changed after creation (no adding, removing, or modifying elements).
# - Indexable & Sliceable → Access elements by index (t[0]) or slice (t[1:3]).
# - Allow duplicates → Same value can appear multiple times.
# - Heterogeneous → Can store mixed data types (numbers, strings, objects).
# - Hashable (if elements are hashable) → Tuples can be used as dictionary keys or set elements.
```

```python
tup_1 = (1,2,3, 'apple', 'banana', 'cherry')
print(tup_1)
print(type(tup_1))
#ordered indexable
# print(tup_1[0])
#Negative Indexing
# print(tup_1[-2])
#Range of Indexes
# tup_ran = tup_1[2:5]#slicing
# print(tup_ran)
# tup_ran = tup_1[:5]
# print(tup_ran)
#update value
tup_1_list = list(tup_1) 
tup_1_list.append('mango')
print(tup_1_list)
tup_1=tuple(tup_1_list)
print(tup_1)
print(type(tup_1))
#Another way
tup_2 =('berry',)
tup_1+=tup_2
print(tup_1)
#Unpack Tuple
fruit = ('apple', 'banana', 'cherry')
fruit_1,fruit_2,fruit_3 = fruit
print(fruit_1,fruit_2,fruit_3)
(*a,b)=fruit
print(a)
```
