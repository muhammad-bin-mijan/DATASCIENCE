```python
#Set
#Duplicate values not allowed
#immutable and unordered, unindexable
my_set = {1, "apple",1,2,20.123}#mixed type allowed
print(my_set)
print(type(my_set))
print(len(my_set))
```

```python
#looping through a set
for item in my_set:
    print(item)

print("banana" in my_set)
```

```python
#Add items in a set
# You cannot change a item in a set but can add items in a set
#add() method 
my_set = {'apple','banana','cherry'}
my_set.add('donut')
my_set
#update
# add any iterable to set
your_set = {"ant", "alu", "pear"}#set
my_set.update(your_set)
my_set
my_set.update([1,2,3,4])#list
my_set
#union()
this_set = {1,2,3,4}
that_set = {2,3,4,5,6,7}
my_set = this_set.union(that_set)#returns a new set
print(my_set)
```

```python
#remove specified item
my_set = {'1', '2', '3',123}
# my_set.remove(123)
my_set.discard(12)#If the element I want to discard doesn't exist, no error will be raised.
my_set.pop()#random item remove
my_set.clear()
my_set
```

```python
#looping through set
my_set = {'apple','banana','cherry'}
for item in my_set:
    print(item)
```

```python
#set methods
#1.copy() method (Returns a set)
pen_set = {"Matador", "Link", "Retro"}
pen_set_copy = pen_set.copy()
print(pen_set_copy)

#intersection() returns a set that is the intersection of two other sets
set_1 = {1,2,3,4,5}
set_2 = {2,3,4,5}
set_3 =set_1.intersection(set_2)
print(set_3)
```
